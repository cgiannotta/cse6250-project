```python
import pandas as pd
```

```python
df_annotations = pd.read_csv('data_load/annotations.csv')
df_annotations.head()
```

```python
df_notes = pd.read_csv('data_load/NOTEEVENTS.csv')
```

```python
df_notes.head()
```

```python
df_notes.CATEGORY.unique()
```

```python
df_discharge = df_notes[df_notes.CATEGORY == 'Discharge summary']
```

```python
df_new = pd.merge(df_annotations, df_notes, how='inner', left_on=['subject.id','Hospital.Admission.ID'], right_on=['SUBJECT_ID','HADM_ID'])
```

```python
df_new.head()
```

```python
# Sanity check on the join
df_new[df_new['Hospital.Admission.ID'] != df_new['HADM_ID']]
```

```python
# Sanity check on the join
df_new[df_new['subject.id'] != df_new['SUBJECT_ID']]
```

```python
df_new = df_new.drop(['Hospital.Admission.ID','subject.id'], axis=1)
```

```python
df_new.head()
```

```python
df_new.to_csv('data_load/annotated_discharge.csv', index=False)
```
